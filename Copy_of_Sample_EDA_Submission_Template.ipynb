Write the summary here within 500-600 words.

This exploratory data analysis (EDA) project investigates the Airbnb listings in New York City for the year 2019. The dataset contains diverse information about listings, including pricing, location, availability, room types, and host details. The primary goal of this analysis is to derive actionable insights about user behavior, host activity, and geographic distribution of listings, which can potentially help in understanding market trends and informing future business or policy decisions.

1. Data Overview and Cleaning
The dataset comprises 48,895 entries and 16 columns, each representing various attributes like name, host_id, neighbourhood, room_type, price, minimum_nights, and more. Initial inspection revealed the presence of missing values, especially in the reviews_per_month column. These were handled using the median imputation method, ensuring that the dataset remained consistent without distorting the distribution of reviews.

Redundant or irrelevant columns such as id, host_name, and last_review were dropped to streamline the analysis. Furthermore, categorical columns like neighbourhood_group, room_type, and neighbourhood were standardized and validated for consistency.

2. Univariate Analysis
A detailed univariate analysis revealed the following:

Neighbourhood Group Distribution: The majority of listings are concentrated in Manhattan and Brooklyn, indicating these are the most preferred areas for Airbnb hosts and travelers.

Room Type Preference: Entire home/apartment is the most listed room type, followed by private rooms, suggesting a strong market for private and exclusive stays.

Price Distribution: The price distribution is right-skewed, with most listings priced under $200. Outliers with exorbitantly high prices were observed and noted for further investigation.

Bar plots, histograms, and pie charts were used extensively here to display frequency distributions and categorical proportions.

3. Bivariate and Multivariate Analysis
The correlation between numerical variables highlighted a few interesting patterns:

Price vs Room Type: Entire homes and apartments are generally priced higher compared to private or shared rooms.

Price vs Neighbourhood Group: Listings in Manhattan command the highest average price, followed by Brooklyn, indicating demand and possibly quality of listings in these areas.

Availability vs Reviews per Month: Highly available listings with consistent reviews indicate active and trusted hosts.

A heatmap of correlations and scatter plots was employed to depict relationships between variables like price, minimum_nights, availability_365, and reviews_per_month.

4. Geospatial Insights
Using latitude and longitude coordinates, a scatter plot of listings across NYC was generated. Manhattan showed the highest density of listings, particularly in areas like Midtown and Lower Manhattan. These findings align with the popularity of Manhattan as a tourist destination.

A geographic heatmap (or point scatter) was particularly effective in visually demonstrating this clustering and offering a sense of spatial saturation.

5. Outlier Detection and Filtering
Several listings were identified as outliers, particularly in terms of price (above $1000) and minimum nights (more than 365). These were either flagged or removed based on logical assumptions and domain knowledge to avoid skewing the analysis.

Box plots and violin plots helped in clearly identifying these outliers.

6. Key Findings and Business Insights
Manhattan and Brooklyn dominate the Airbnb landscape both in terms of quantity and price.

Entire homes/apartments are in high demand and can generate significantly more revenue than shared accommodations.

Hosts with frequent reviews and high availability are more likely to be experienced and reliable, which could be useful for marketing or platform policy decisions.

Price optimization should consider both room type and location, as they are strongly correlated with listing value.
